## Combined Search, Clustering & Wordcloud

```python
def analyze_query(query , k=5):
  query_embedding = model.encode([query])
  faiss.normalize_L2(query_embedding)
  D,I = index.search(query_embedding , k)

  results = []
  for score,idx in zip(D[0],I[0]):
    title = df.iloc[idx]["title"]
    abstract = df.iloc[idx]["abstract"]

    summary = summarizer(abstract , max_length=120 , min_length=40 , do_sample=False)
    keywords = kw_model.extract_keywords(abstract , keyphrase_ngram_range=(1,3) , stop_words="english" , top_n=5)

    results.append({
        "title": title,
        "score": float(score),
        "summary": summary[0]["summary_text"],
        "keywords": [k for k,v in keywords]
    })

  return results
```

```python
res = analyze_query("deep learning for medical image analysis")
for r in res:
  print(r["title"])
  print(r["score"])
  print(r["summary"])
  print(r["keywords"])
  print()
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

kmeans = KMeans(n_clusters=8 , random_state=42 , n_init=10)
clusters = kmeans.fit_predict(embedding)

df["cluster"] = clusters
df["cluster"].value_counts()
```

```python
pca = PCA(n_components=2)
reduced = pca.fit_transform(embedding)

plt.figure(figsize=(9,6))
plt.scatter(reduced[:,0] , reduced[:,1] , c=clusters , cmap="tab10" , s=5)
plt.title("paper clusters")
plt.show()
```
